```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib inline
from matplotlib import animation, rc
from IPython.display import HTML

# Set up figure and animation
fig, ax = plt.subplots()

ax.set_facecolor('xkcd:black')

# Set up the sun and planets as circles with colors
sun = plt.Circle((0, 0), 1.0, color='yellow')
earth = plt.Circle((7, 0), 0.4, color='blue')
rocket = plt.Circle((7, 0), 0.1, color='red')
mars = plt.Circle ((8, 0), 0.3, color='brown')
venus = plt.Circle ((6,0), 0.3, color = 'white')


# Add the sun and planets to the plot
ax.add_patch(sun)
ax.add_patch(earth)
ax.add_patch(rocket)
ax.add_patch(mars)
ax.add_patch(venus)

# Set axis limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Set up the animation function
def animate(i):
    # Animate the rocket orbiting around the earth
    lineofx=np.linspace(2,i,num=100)
    mars.center = (2*np.cos(i*10/1), 1*np.sin(i*10/1))
    venus.center = (5*np.cos(i*25/5), 3*np.sin(i*25/5))
    rocket.center=(7-i, 0)
     
    return sun, earth, rocket, mars, venus

# Run the animation
ani = FuncAnimation(fig, animate, frames=np.linspace(0, 5, 500), interval=100,
                    repeat=False)

    
HTML(ani.to_html5_video())

```
